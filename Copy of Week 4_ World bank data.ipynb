```python
def preprocess_df(df, value_name):
    """ remove missing values and put years in one column

    Parameters
    ----------
    df: dataframe
        the data that needs to be preprocessed

    value_name: string
        the name of the column that will contain the year's data

    Return
    ------
    preprocessed dataframe
    """
    years = [str(y) for y in range(1960, 2017)]

    # remove useless columns
    df.drop(['Indicator Name', 'Indicator Code'], axis=1, inplace=True)

    # remove countries with missing value
    df.dropna(axis=0, inplace=True)

    # melt the dataframe to have years in one columns
    df = pd.melt(df,
                 id_vars='Country Code',
                 value_vars=years,
                 var_name='Year',
                 value_name=value_name)

    return df
```

```python
country = country[['Country Code', 'Region']]
population = preprocess_df(population, 'Population')
fertility_rate = preprocess_df(fertility_rate, 'Fertility Rate')
life_expectancy = preprocess_df(life_expectency, 'Life Expectancy')
```

```python
metadata = df[['Country Code', 'Region']].dropna()

G = nx.Graph()

nodes = metadata['Country Code'].tolist()
G.add_nodes_from(nodes)

for region in metadata['Region'].unique():
    countries_in_region = metadata[metadata['Region'] == region]['Country Code'].tolist()
    for i in range(len(countries_in_region)):
        for j in range(i + 1, len(countries_in_region)):
            G.add_edge(countries_in_region[i], countries_in_region[j])
```

```python
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=300, node_color='lightgreen', font_size=8, font_weight='bold', edge_color='gray')

plt.title('Graph of Countries Connected by Region')
plt.show()
```
